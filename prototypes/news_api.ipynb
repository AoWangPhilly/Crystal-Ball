```python
import os

from newsapi import NewsApiClient
import pandas as pd
```

```python
api = NewsApiClient(api_key=os.getenv("NEWS_API_KEY"))
```

```python
# /v2/top-headlines
everything = api.get_everything(
    q="twitter",
    language="en",
    from_param='2022-11-21',
    to='2022-11-25',
)
```

```python
df = pd.DataFrame(everything["articles"])
```
